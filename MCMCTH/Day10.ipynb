```python
import matplotlib.pyplot as plt
import math
import numpy as np

x_data = np.arange(0,10,0.1)
y_data = [math.exp(x) for x in x_data]

plt.figure(figsize=(6,4))
plt.plot(x_data, y_data)

plt.show()
```
